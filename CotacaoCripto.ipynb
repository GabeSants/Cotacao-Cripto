```python
import requests
import matplotlib.pyplot as plt

def get_crypto_price():
  url = "https://api.coingecko.com/api/v3/simple/price"
  params = {
      "ids": "bitcoin,ethereum,litecoin,dogecoin,solana,mantra,xrp", # criptomoedas
      "vs_currencies": "usd,brl" # moedas de conversão
  }

  response = requests.get(url, params=params)

  if response.status_code == 200:
    data = response.json()

    # Exibir os preços no console
    print(f"Bitcoin: {data['bitcoin']['usd']} | R$ {data['bitcoin']['brl']}")
    print(f"Ethereum: {data['ethereum']['usd']} | R$ {data['ethereum']['brl']}")
    print(f"Litecoin: {data['litecoin']['usd']} | R$ {data['litecoin']['brl']}")
    print(f"Dogecoin: {data['dogecoin']['usd']} | R$ {data['dogecoin']['brl']}")
    print(f"Solana: {data['solana']['usd']} | R$ {data['solana']['brl']}")

    return data # Retorna os dados para serem usados depois

  else:
    print("Erro ao buscar os preços das criptomoedas")
    return None # Retorna None em caso de erro


data = get_crypto_price() # pegando os preços

if data:
  print("Dados carregados com sucesso!")
else:
  print("Erro: Dados não carregados")
```

```python
def plot_crypto_price(data):
    crypto_names = list(data.keys())
    prices_usd = [data[coin]['usd'] for coin in crypto_names]
    prices_brl = [data[coin]['brl'] for coin in crypto_names]

    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Criar gráfico
    axs[0].plot(crypto_names, prices_usd, marker='o', linestyle='-', color='blue', linewidth=2, label='Preço (USD)')
    axs[0].set_title('Preços de Criptomoedas', fontsize=14)
    axs[0].set_xlabel('Criptomoedas', fontsize=12)
    axs[0].set_ylabel('Preço (USD)', fontsize=12)
    axs[0].tick_params(axis='x', rotation=25)
    axs[0].legend()
    axs[0].grid(alpha=0.3)

    axs[1].plot(crypto_names, prices_brl, marker='s', linestyle='-', color='green', linewidth=2, label='Preço (BRL)')
    axs[1].set_title('Preços de Criptomoedas', fontsize=14)
    axs[1].set_xlabel('Criptomoedas', fontsize=12)
    axs[1].set_ylabel('Preço (BRL)', fontsize=12)
    axs[1].tick_params(axis='x', rotation=25)
    axs[1].legend()
    axs[1].grid(alpha=0.3)

    # Ajustar layout
    plt.tight_layout()
    # Exibir gráfico
    plt.show()

plot_crypto_price(data) # Gerar o gráfico
```
